```python
# Read data from the file and handle null values
import pandas as pd
df = pd.read_csv('raw_data_epl_stats.csv')
df.head(10)
# df['MarketAvgDraw'].unique()
```

```python
# Rename column names to standardize them into lower case and replace space with underscore (Can't replace all as there're too many columns)

#Below is the manual rename (Bad practice)
# df.rename(columns={'MatchID': 'match_id', 'Season': 'season', 'MatchWeek': 'match_week', 'Date': 'date', 'Time': 'time', 'HomeTeam': 'home_team', 'AwayTeam': 'away_team', 'FullTimeHomeTeamGoals': 'fulltime_home_team_goals', 'FullTimeAwayTeamGoals': 'fulltime_away_team_goals', 'FullTimeResult': 'fulltime_result'})

# Below is automatic rename (Best Practice)
df.columns = df.columns.str.lower()

# In case within the string, there's a space for now reason, we can standardize by replacing them with '_'
# df.columns = df.columns.str.replace('','_')

# Show the current column of the data file after change
df.columns
```

```python
# Derive new insights
# Shooting Accuracy for Home Team
import numpy as np

# Fill NaN values in relevant columns with 0
df['hometeamshots'].fillna(0, inplace=True)
df['hometeamshotsontarget'].fillna(0, inplace=True)

# Convert relevant columns to NumPy arrays for faster calculations
hometeamshots = df['hometeamshots'].to_numpy()
hometeamshotsontarget = df['hometeamshotsontarget'].to_numpy()

# Calculate Shooting Accuracy using NumPy and handle division by zero
hometeam_shooting_accuracy = np.zeros_like(hometeamshots, dtype=float)  # Initialize array with zeros

# Only calculate where shots are greater than zero
valid_indices = hometeamshots > 0
hometeam_shooting_accuracy[valid_indices] = (hometeamshotsontarget[valid_indices] / hometeamshots[valid_indices]) * 100 

# Adding the calculated accuracy back to the DataFrame
df['hometeam_shooting_accuracy'] = hometeam_shooting_accuracy
```

```python
df
```

```python
# Shooting Accuracy for Away Team
import numpy as np

# Fill NaN values in relevant columns with 0
df['awayteamshots'].fillna(0, inplace=True)
df['awayteamshotsontarget'].fillna(0, inplace=True)

# Convert Away Team shots column to NumPy arrays for faster calculations
awayteamshots = df['awayteamshots'].to_numpy()
awayteamshotsontarget = df['awayteamshotsontarget'].to_numpy()

# Calculate Shooting Accuracy using NumPy and handle division by zero
awayteam_shooting_accuracy = np.zeros_like(awayteamshots, dtype=float)  # Initialize array with zeros

# Only calculate where shots are greater than zero
valid_indices_awayteam = awayteamshots > 0
awayteam_shooting_accuracy[valid_indices_awayteam] = (awayteamshotsontarget[valid_indices_awayteam] / awayteamshots[valid_indices_awayteam]) * 100 

# Adding the calculated accuracy back to the DataFrame
df['awayteam_shooting_accuracy'] = awayteam_shooting_accuracy
```

```python
df
```

```python
# Home/Away Goal Difference
import numpy as np

# Fill NaN values in goals columns with 0
df['fulltimehometeamgoals'].fillna(0, inplace=True)
df['fulltimeawayteamgoals'].fillna(0, inplace=True)

# Convert Full time home/away team goals column to NumPy arrays for faster calculations
fulltimehometeamgoals = df['fulltimehometeamgoals'].to_numpy()
fulltimeawayteamgoals = df['fulltimeawayteamgoals'].to_numpy()

# Home Goal Difference Formula
df['goal_difference_home'] = fulltimehometeamgoals - fulltimeawayteamgoals
```

```python
df
```

```python
# Away Goal Difference Formula
df['goal_difference_away'] = fulltimeawayteamgoals - fulltimehometeamgoals
```

```python
df
```

```python
import numpy as np

# Calculate Shots on Target Ratio for Home Team
home_shots_on_target_ratio = np.zeros_like(df['hometeamshots'], dtype=float)  # Initialize with zeros
valid_home_indices = df['hometeamshots'] > 0  # Check for valid total shots (non-zero)

# Calculate the ratio where total shots are valid
home_shots_on_target_ratio[valid_home_indices] = (df['hometeamshotsontarget'][valid_home_indices] / df['hometeamshots'][valid_home_indices])

df['home_shots_on_target_ratio'] = home_shots_on_target_ratio
```

```python
df
```

```python
import numpy as np

# Calculate Shots on Target Ratio for Away Team
away_shots_on_target_ratio = np.zeros_like(df['awayteamshots'], dtype=float)  # Initialize with zeros
valid_indices_awayteam = df['awayteamshots'] > 0  # Check for valid total shots (non-zero)

# Calculate the ratio where total shots are valid
away_shots_on_target_ratio[valid_indices_awayteam] = (df['awayteamshotsontarget'][valid_indices_awayteam] / df['awayteamshots'][valid_indices_awayteam])

df['away_shots_on_target_ratio'] = away_shots_on_target_ratio
```

```python
df
```

```python
# Calculate Home Team Performance Index (HPI)
df['home_performance_index'] = (fulltimehometeamgoals + home_shots_on_target_ratio) / 2
```

```python
df
```

```python
# Calculate Away Team Performance Index (HPI)
df['away_performance_index'] = (fulltimeawayteamgoals + away_shots_on_target_ratio) / 2
```
